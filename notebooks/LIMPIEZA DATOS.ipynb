```python
#CONVERTIMOS LAS LOCALIZACIONES EN PROVINCIAS

import pandas as pd
import unicodedata

def normalizar(texto):
    if pd.isna(texto):
        return ""
    texto = texto.lower().strip()
    texto = unicodedata.normalize('NFKD', texto)
    texto = ''.join([c for c in texto if not unicodedata.combining(c)])
    return texto

diccionario = pd.read_excel("C:\\Users\\josef\\Desktop\\BOOTCAMP DATA\\PROYECTO FINAL\\diccionario25.xlsx")  # Diccionario descargado de INE

municipio_col = 'MUN'    
provincia_col = 'PRO'

municipio_a_provincia = dict(
    zip(
        diccionario[municipio_col].apply(normalizar),
        diccionario[provincia_col]
    )
)


data_preparada['provincia'] = data_preparada['ubicacion'].apply(lambda x: municipio_a_provincia.get(normalizar(x)))

no_mapeados = data_preparada[data_preparada['provincia'].isna()]['ubicacion'].unique()
print("Municipios no mapeados:", no_mapeados)
```

```python
# 4. FUNCIÓN GENERAL PARA APLICAR NORMALIZACIÓN
FUNCIONES_NORMALIZACION = {
    'apple': normalizar_modelo_apple,
    'samsung': normalizar_modelo_samsung,
    "xiaomi": normalizar_modelo_xiaomi,
    "huawei": normalizar_modelo_huawei,
    "honor": normalizar_modelo_honor,
    "realme":normalizar_modelo_realme,
    "oneplus":normalizar_modelo_oneplus,
    "google": normalizar_modelo_google,
    "vivo": normalizar_modelo_vivo,
    "motorola":normalizar_modelo_motorola,
    "nokia":normalizar_modelo_nokia,
    "sony": normalizar_modelo_sony,
    "asus":normalizar_modelo_asus,
    "zte":normalizar_modelo_zte,
    "tcl": normalizar_modelo_tcl,
    "alcatel":normalizar_modelo_alcatel,
    "nothing phone":normalizar_modelo_nothing,
    "oppo": normalizar_modelo_oppo,
}

def normalizar_modelo_por_marca(row):
    """
    Aplica la función de normalización específica según la marca
    """
    marca = row['marca']
    modelo = row['modelo']
    
    # Verificar valores nulos
    if pd.isna(marca) or pd.isna(modelo):
        return modelo
    
    # Buscar función específica para la marca
    funcion = FUNCIONES_NORMALIZACION.get(marca.lower())
    
    if funcion:
        return funcion(modelo)
    else:
        # Normalización básica si no hay función específica
        return str(modelo).lower().strip()

# 5. COPIAR EL DATAFRAME Y APLICAR LA NORMALIZACIÓN
print("🔄 Aplicando normalización de modelos...")

# Crear nueva copia para no modificar el original
data_modelos_normalizada = data_preparada.copy()

# Aplicar normalización a nueva columna
data_modelos_normalizada['modelo_normalizado'] = data_modelos_normalizada.apply(
    normalizar_modelo_por_marca, axis=1
)

# 6. VERIFICAR RESULTADOS
print("✅ Normalización completada.")
print(f"📊 Filas procesadas: {len(data_modelos_normalizada)}")
print(data_modelos_normalizada[['marca', 'modelo', 'modelo_normalizado']].head(10))

```
